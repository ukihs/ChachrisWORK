# Manga that End

```python

#มังงะเรื่องไหรจบแล้ว/ยังไม่จบและทั้งหมดมีกี่เรื่องกี่เล่ม
info = input('Ongoing or End: ')
if info == 'Ongoing' :
  ongoings = data['Manga'].where(data['Volumes'] == 'Ongoing')
  volume = data['Volumes'].where(data['Volumes'] == 'Ongoing')
  list1 = {'Manga': ongoings.dropna(),'Volumes': volume.dropna()}
  dfmv = pd.DataFrame(list1)
  print(dfmv)
  number = ongoings.count()
  print('Numbers of Ongoing Manga is :',number)
else :
  end = data['Manga'].where(data['Volumes'] != 'Ongoing')
  volume = data['Volumes'].where(data['Volumes'] != 'Ongoing')
  list2 = {'Manga': end.dropna(),'Volumes': volume.dropna()}
  dfmv = pd.DataFrame(list2)
  print(dfmv)
  ending = end.count()
  print('Numbers of Manga that End is :',ending)
```

# Mangas that have hundred Chap.

```python
def change(int):
  data['Chapters'] = i.astype(int)
#เรื่องไหนที่มีจำนวนตอน 100 ขึ้นไปหรือต่ำกว่า
hdvar = input('More Than Hund or Lower Than Hund(More,Lower) : ')
if hdvar == 'More':
  endhd = data['Chapters'].where(data['Chapters'] != 'Ongoing' )
  i = endhd.dropna()
  change(int)
  name = data['Manga'].where(data['Chapters'] >=100)
  dfname = name.dropna()
  dfhd = data['Chapters'].where(data['Chapters'] >= 100 )
  dflist = {'Manga': dfname,'Chapters': dfhd.dropna()}
  dfmore = pd.DataFrame(dflist)
  print(dfmore.dropna())
  number = dfhd.count()
  print('Number of Mangas Than have chapters More than Hund :',number)
else  :
  endhd = data['Chapters'].where(data['Chapters'] != 'Ongoing' )
  i = endhd.dropna()
  change(int)
  name = data['Manga'].where(data['Chapters'] <100)
  dfname = name.dropna()
  dfhd = data['Chapters'].where(data['Chapters'] < 100 )
  dflist = {'Manga': dfname,'Chapters': dfhd.dropna()}
  dflower = pd.DataFrame(dflist)
  print(dflower.dropna())
  number = dfhd.count()
  print('Number of Mangas Than have chapters lower than Hund :',number)
```

#Demographic for Manga


```python
#กลุ่มประชากรที่เหมาะแก่การดูตรงกับเรื่องอะไรบ้าง
demoinfo = input('Demographic(Shounen, Seinen, Shoujo, Josei): ')
if demoinfo == 'Shounen' :
  shounen = data['Manga'].where(data['Demographic'] == 'Shounen')
  demo = data['Demographic'].where(data['Demographic'] == 'Shounen')
  demolist = {'Manga': shounen.dropna(),'Demographic': demo.dropna()}
  dfsho = pd.DataFrame(demolist)
  print(dfsho)
  number = shounen.count()
  print('Numbers of Shounen Manga is :',number)
elif demoinfo == 'Seinen' :
  seinen = data['Manga'].where(data['Demographic'] == 'Seinen')
  demo2 = data['Demographic'].where(data['Demographic'] == 'Seinen')
  demolist = {'Manga': seinen.dropna(),'Demographic': demo2.dropna()}
  dfsei = pd.DataFrame(demolist)
  print(dfsei)
  number = seinen.count()
  print('Numbers of Seinen Manga is :',number)
elif demoinfo == 'Shoujo' :
  shoujo = data['Manga'].where(data['Demographic'] == 'Shoujo')
  demo3 = data['Demographic'].where(data['Demographic'] == 'Shoujo')
  demolist = {'Manga': shoujo.dropna(),'Demographic': demo3.dropna()}
  dfjo = pd.DataFrame(demolist)
  print(dfjo)
  number = shoujo.count()
  print('Numbers of Shoujo Manga is :',number)
elif demoinfo == 'Josei':
  josei = data['Manga'].where(data['Demographic'] == 'Josei')
  demo4 = data['Demographic'].where(data['Demographic'] == 'Josei')
  demolist = {'Manga': josei.dropna(),'Demographic': demo4.dropna()}
  dfjos = pd.DataFrame(demolist)
  print(dfjos)
  number = josei.count()
  print('Numbers of Josei Manga is :',number)
else :
  none = data['Manga'].where(data['Demographic'] == 'None')
  demo5 = data['Demographic'].where(data['Demographic'] == 'None')
  demolist = {'Manga': none.dropna(),'Demographic': demo5.dropna()}
  dfno = pd.DataFrame(demolist)
  print(dfno)
  number = none.count()
  print('Numbers of Manga that not fit in demographic is :',number)

test = data['demographic']
te1 = test.count()
print(te1)
```
